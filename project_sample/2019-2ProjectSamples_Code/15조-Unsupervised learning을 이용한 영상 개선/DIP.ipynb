***deep image prior(DIP)?***

> Deep image prior는 이미지 자체 이외의 사전 training data 없이 주어진 이미지를 개선시키는데 CNN 을 사용합니다. 이 프로젝트는 현재 영상복원 문제에 있어서 다루어지는 서로 매우 다른 두 방향의 연구들을 이어준 프로젝트입니다. (model-based optimization vs deep learning)

cf)
![table](https://drive.google.com/uc?export=view&id=1bySgqgP9-mnmMofpdE2xawG2Ycv3jDWj)


DIP는 대다수의 deep supervised learning 방식들과 달리, **"네트워크 구조 그 자체" 를 inverse prolem의 regularizer로 사용하였다고 합니다.** 즉, denoising, super-resolution, inpainting과 같은 inverse problem을 해결하기 전에 CNN이 random하게 초기화되어 사용됩니다. 
( 즉, CNN의 구조 자체가 내재적인 prior라서 CNN이 표현 가능한 parametric function space가 일반적이지 않다는 가정에서 출발합니다.)

이를 바탕으로 deep-image-prior는 우리가 모르는 latent image x를 CNN output으로 놓되, z라는 고정된 random vector에 대해서 다음과 같은 objective function에 대한 optimization 문제를 풀게 됩니다.

![수식](https://drive.google.com/uc?export=view&id=1yg1fnC8US6J2WhD-hJNUr57M2sgTM_y-)


- 어떤 noise가 더해진 이미지를 주고 CNN을 사용하여 이런 이미지를 아무리 fitting하려 해도, 해당 CNN으로 대표되는 parametic function space의 함수들로는 성질이 좋지 않은 noise를 표현할 수가 없기에 natural image만을 fitting하여 깨끗한 결과를 내보낸다는 개념입니다.

또한 CNN이 이미지 한장 한장에 대해 optimization문제를 풀어 theta 값을 구합니다. 이는 model-based optimization방식과 비슷한 점인데, model-based optimization이 그러하듯 DIP도 target data 없이, unsupervised 방식으로 영상복원 문제를 풀 수 있게 됩니다.



***concept detail(optimization 과정)***
![과정](https://drive.google.com/uc?export=view&id=1rb_y5D1F6tj-EudN5GRCNR-O1UVuAHi9)


임의 가중치 0에서 시작하여, data term equation을 최소화 하기 위해 가중치를 반복적으로 업데이트합니다. 매 iteration마다 가중치 theta는 영상 x에 mapping되며, 이미지 x는 E(x, x0)를 계산하는데 사용됩니다. 또한 loss w.r.t의 기울기는 가중치가 계산되어 parameter들을 업데이트 하는 데 사용되어집니다.
